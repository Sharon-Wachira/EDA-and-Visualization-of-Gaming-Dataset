```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset
df=pd.read_csv(r'C:\Users\Gamers.csv')
df 
```

```python
# Set up plot aesthetics
sns.set(style="whitegrid")

# Bar plot to visualize which genre has the most games
plt.figure(figsize=(10, 6))
genre_count = df['Genre'].value_counts()
sns.barplot(x=genre_count.index, y=genre_count.values, color='red') 
plt.title('Number of Games per Genre', fontsize=16)
plt.xlabel('Genre', fontsize=14)
plt.ylabel('Number of Games', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Count the number of games released per year
year_count = df['Year'].value_counts().sort_index()

# Find the year with the most games released
max_year = year_count.idxmax()
max_count = year_count.max()

# Plotting the data
plt.figure(figsize=(10, 6))
sns.lineplot(x=year_count.index, y=year_count.values, marker="o", color='brown')
plt.title('Number of Games Released per Year')
plt.xlabel('Year')
plt.ylabel('Number of Games')

# Annotate the year with the most games released
plt.annotate(f'Most Games: {max_year} ({max_count})', 
             xy=(max_year, max_count), 
             xytext=(max_year, max_count + 10), 
             arrowprops=dict(arrowstyle='->', color='black'),
             fontsize=10,
             color='black',
             ha='center')

plt.show()
```

```python
# Group the data by Genre and sum the JP_Sales
jp_sales_by_genre = df.groupby('Genre')['JP_Sales'].sum().sort_values(ascending=False)

# Plotting the data
plt.figure(figsize=(10, 6))
sns.barplot(x=jp_sales_by_genre.index, y=jp_sales_by_genre.values, color='maroon')  
plt.title('Total Sales by Genre in Japan', fontsize=16)  
plt.xlabel('Genre', fontsize=12)  
plt.ylabel('Total Sales in Japan (in millions)', fontsize=12)  
plt.xticks(rotation=45)  
plt.show()  
```

```python
 #Count the number of games released each year
games_per_year = df['Year'].value_counts()

# Identify the year with the most games released
max_games_year = games_per_year.idxmax()
max_games_count = games_per_year.max()
print(f"The year with the most games released is {max_games_year} with a total of {max_games_count} games.")

# Identify the year with the least games released
min_games_year = games_per_year.idxmin()
min_games_count = games_per_year.min()
print(f"The year with the least games released is {min_games_year} with a total of {min_games_count} games.")
```

```python
# Sales trend comparison between countries/continent and global sales
sales_comparison = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].sum()
print(f"Sales comparison by region:\n{sales_comparison}")
```

i)Best- selling video game in North America

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_na_sales = df[['Name', 'NA_Sales']].sort_values(by='NA_Sales', ascending=False).head(10)

# Plotting the data
plt.figure(figsize=(12, 6))
sns.barplot(x='NA_Sales', y='Name', data=top_na_sales, color='maroon')  # Set bar color to maroon

plt.title('Top 10 Best-Selling Video Games in North America', fontsize=16)
plt.xlabel('Sales in North America (in millions)', fontsize=14)
plt.ylabel('Video Game Title', fontsize=14)
plt.tight_layout()
plt.show()
```

 ii)	Country which exhibits a consistent sales trend for the past thirty-two years since 1985.

```python
# Preparing the data for sales trend since 1985
df_filtered = df[df['Year'] >= 1985]  

# Grouping sales by year and country/region
sales_by_year = df_filtered.groupby('Year')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

plt.figure(figsize=(10, 6))
for region in ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']:
    plt.plot(sales_by_year.index, sales_by_year[region], label=region)

plt.title('Consistent Sales Trend by Region (1985 - Present)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Sales (in millions)', fontsize=12)
plt.legend(title='Region')
plt.grid(True)
plt.show()

# Calculate standard deviation of sales for each region (lower = more consistent)
sales_std_dev = sales_by_year.std()
print(sales_std_dev)

# Region with the most consistent sales trend
most_consistent_region = sales_std_dev.idxmin()
print(f"The region with the most consistent sales trend is: {most_consistent_region}")
```
